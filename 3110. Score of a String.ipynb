```python
class Solution(object):
    def scoreOfString(self, s):
        """
        :type s: str
        :rtype: int
        """
        result = 0 
        for i in range (len(s) - 1):
            difference = abs(ord(s[i]) - ord(s[i + 1]))
            result += difference
        return result
        
```

```python
s = "hello"
solution = Solution()
print(solution.scoreOfString(s))
```

```python
s = "zaz"
solution = Solution()
print(solution.scoreOfString(s))
```
