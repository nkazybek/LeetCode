```python
class Solution(object):
    def findClosest(self, x, y, z):
        """
        :type x: int
        :type y: int
        :type z: int
        :rtype: int
        """
        distanceX = abs(x - z)
        distanceY = abs(y - z)
        
        if distanceX > distanceY:
            return 2
        elif distanceX < distanceY:
            return 1
        else:
            return 0
```

```python
x = 2
y = 7
z = 4
sol = Solution()
sol.findClosest(x, y, z)
```
