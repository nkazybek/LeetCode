```python
#Leetcode #1

from typing import List

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        for i in range(len(nums) - 1):
            for j in range(i + 1, len(nums)):
                if nums[i] + nums[j] == target:
                    ll = list([i])
                    ll.append(j)
                    return ll

nums = [2,7,11,15]
target = 9
solution = Solution()
print(solution.twoSum(nums, target))
```

```python
#Leetcode #283

class Solution:
    def moveZeroes(self, nums: List[int]) -> None:
        for i in range(len(nums)):
            if nums[i] == 0:
                for j in range(i + 1, len(nums)):
                    if nums[j] != 0:
                        nums[i], nums[j] = nums[j], nums[i]
                        break
        print(nums)

nums = [0]
solution = Solution()
solution.moveZeroes(nums)
```

```python
#Leetcode #448

class Solution:
    def findDisappearedNumbers(self, nums: List[int]) -> List[int]:
        my_list = []
        for i in range(1, len(nums) + 1):
            if not nums.count(i):
                my_list.append(i)
        return my_list
                
nums = [4,3,2,7,8,2,3,1]
solution = Solution()
solution.findDisappearedNumbers(nums)
```

```python
#Leetcode #448_2

class Solution:
    def findDisappearedNumbers(self, nums: List[int]) -> List[int]:
        temp_set = set(nums)
        temp_set2 = set()
        for i in range(1, len(nums) + 1):
            temp_set2.add(i)
        return temp_set ^ temp_set2
        
nums = [4,3,2,7,8,2,3,1]
solution = Solution()
solution.findDisappearedNumbers(nums)
```

```python
#Leetcode #206
# Definition for singly-linked list.

from typing import Optional

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
        
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        
        self.head = head
        prev = None
        current = self.head
        
        while current:
            nxt = current.next
            current.next = prev
            prev = current
            current = nxt
            self.head = prev
        
        return current

        
head = [1,2,3,4,5]
solution = Solution()
solution.reverseList(head)
```

```python
#Leetcode #169
from typing import List

class Solution:
    def majorityElement(self, nums: List[int]) -> int:
        max_n = 0
        max_count = 0
        count = 1
        if len(nums) == 1:
            return nums[0]
        
        nums = sorted(nums)
        for i in range(1, len(nums)):
            if nums[i] == nums[i - 1]:
                count += 1
                if i == len(nums) - 1:
                    if max_count < count:
                        max_count = count
                        max_n = nums[i]
                
            elif nums[i] != nums[i - 1]:
                if max_count < count:
                    max_count = count
                    max_n = nums[i - 1]
                    count = 1
        return max_n
        
solution = Solution()
nums = [2]
solution.majorityElement(nums)
```

```python
#Leetcode #66

class Solution:
    def plusOne(self, digits: List[int]) -> List[int]:
        number = 0
        for i in range(len(digits)):
            number = number + 10 ** (len(digits) - i - 1) * digits[i]
        
        number += 1
        digits = list(map(int, str(number)))
            
        return digits
        
        
digits = [4,3,2,1]   
solution = Solution()
solution.plusOne(digits)
```

```python
#Leetcode #136

class Solution:
    def singleNumber(self, nums: List[int]) -> int:
        if len(nums) == 1:
            return nums[0]
        nums = sorted(nums)
        for i in range(0,len(nums) - 1, 2):
            if nums[i] != nums[i + 1]:
                return nums[i]
        
        return nums[-1]
        
nums = [-336,513,-560,-481,-174,101,-997,40,-527,-784,-283,-336,513,-560,-481,-174,101,-997,40,-527,-784,-283,354]
solution = Solution()
solution.singleNumber(nums)
```

```python
#Leetcode #189
from typing import List

class Solution:
    def rotate(self, nums: List[int], k: int) -> None:
        if k >= len(nums):
            k = k % len(nums)
        for i in range(k):
            tmp = nums[-1]
            for j in range(len(nums)-1, -1, -1):
                nums[j] = nums[j - 1]
            nums[0] = tmp
                

nums = [1,2,3,4,5,6,7]
k = 3

sol = Solution()
sol.rotate(nums, k)
```
