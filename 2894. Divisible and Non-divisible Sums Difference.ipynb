```python
class Solution(object):
    def differenceOfSums(self, n, m):
        """
        :type n: int
        :type m: int
        :rtype: int
        """
        divisible = 0
        notDivisible = 0
        
        for i in range(1, n + 1):
            if (i % m == 0):
                divisible += i
            else:
                notDivisible +=  i
        return notDivisible - divisible
    
        
```

```python
n = 10
m = 3
solution = Solution()
print(solution.differenceOfSums(n, m))
```

```python
n = 5
m = 6
solution = Solution()
print(solution.differenceOfSums(n, m))
```
