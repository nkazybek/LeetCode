```python
class Solution(object):
    def leftRightDifference(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        answer = []
        for i in range(len(nums)):
            summ = 0
            
            if i == 0:
                for j in range(i + 1, len(nums)):
                    summ += nums[j]
                answer.append(summ)
            
            elif i == (len(nums) - 1):
                for j in range(len(nums) - 1):
                    summ += nums[j]
                answer.append(summ)
            else:
                for j in range(i + 1, len(nums)):
                    summ += nums[j]
                for j in range(i):
                    summ -= nums[j]
                answer.append(abs(summ))
        
        return answer    
```

```python
nums = [10,4,8,3]
sol = Solution()
sol.leftRightDifference(nums)
```
