```python
class Solution(object):
    def concatWithReverse(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        result = []
        length = len(nums) * 2
        for i in range(length):
            if i < len(nums):
                result.append(nums[i])
            else:
                result.append(nums[length - i - 1])
                
        return result
```

```python
nums = [1,2,3]
sol = Solution()
sol.concatWithReverse(nums)
```
