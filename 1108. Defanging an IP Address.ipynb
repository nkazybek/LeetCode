```python
class Solution(object):
    def defangIPaddr(self, address):
        """
        :type address: str
        :rtype: str
        """
        result = ""
        for char in address:
            if char != '.':
                result += char
            else:
                result += "[.]"
                
        return result
```

```python
address = "1.1.1.1"
sol = Solution()
sol.defangIPaddr(address)
```

```python
address = "255.100.50.0"
sol = Solution()
sol.defangIPaddr(address)
```
