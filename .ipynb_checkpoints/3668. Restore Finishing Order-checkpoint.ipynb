```python
class Solution(object):
    def recoverOrder(self, order, friends):
        """
        :type order: List[int]
        :type friends: List[int]
        :rtype: List[int]
        """
        result = [len(friends)]
        for i in order:
            for j in friends:
                if (i == j)
        return 0
```

```python
order = [3,1,2,5,4]
friends = [1,3,4]

solution = Solution()
print(solution.recoverOrder(order, friends))
```
