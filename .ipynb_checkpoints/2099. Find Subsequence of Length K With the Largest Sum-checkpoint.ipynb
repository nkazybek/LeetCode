```python
class Solution(object):
    def maxSubsequence(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: List[int]
        """
        #nums.sort()
        given = nums.copy()
        result = []
        value = []
        #for i in range(len(nums) - k, len(nums), 1):
        #    result.append(nums[i])
            
        
        for i in range(k):
            maxNumber = max(nums);
            result.append(maxNumber)
            for j in range(len(nums)):
                if nums[j] == maxNumber:
                    nums.pop(j)
                    break
        
        for i in range(len(given)):
            j = 0
            while j < len(result):
                if given[i] == result[j]:
                    value.append(result[j])
                    result.pop(j)
                j += 1
            for j in range(len(result)):
                if given[i] == result[j]:
                    value.append(result[j])
        return value
        # 1 2 3 3
        # 0 1 2 3 
        # len 4
        # k 2 
        # 
```

```python
nums = [2,1,3,3]
k = 2
s = Solution()
s.maxSubsequence(nums, k)
```

```python
nums = [-1,-2,3,4]
k = 3
s = Solution()
s.maxSubsequence(nums, k)
```
