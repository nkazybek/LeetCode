```python
class Solution(object):
    def numJewelsInStones(self, jewels, stones):
        """
        :type jewels: str
        :type stones: str
        :rtype: int
        """
        answer = 0
        for jewel in jewels:
            for stone in stones:
                if jewel == stone:
                    answer += 1
        return answer
```

```python
jewels = "aA"
stones = "aAAbbbb"
sol = Solution()
sol.numJewelsInStones(jewels, stones)
```
