```python
class Solution(object):
    def longestCommonPrefix(self, strs):
        """
        :type strs: List[str]
        :rtype: str
        """
        result = ""
        if len(strs) > 0:
            
            first = strs[0] #flower

            boo = True
            iterator = 0
            
            if len(first) > 0:
                prefix = first[0]
            else:
                prefix = ""
                
            while (boo and (iterator < len(first))):
                count = 0

                for word in strs:
                    if word.startswith(prefix):
                        count += 1

                if count != len(strs):
                    boo = False
                else:
                    if (len(first) > iterator):
                        result = prefix;
                        
                        iterator += 1
                        if (len(first) <= iterator):
                            continue
                        else:
                            prefix = prefix + first[iterator]

        return result
        
        
```

```python

strs = ["flower","flow","flight"]

solution = Solution()
solution.longestCommonPrefix(strs)
```
