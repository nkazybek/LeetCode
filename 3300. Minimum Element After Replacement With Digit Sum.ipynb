```python
class Solution(object):
    def minElement(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        answer = []
        for num in nums:
            summ = 0
            while num > 9:
                tmp = num % 10
                num = num // 10
                summ += tmp
            summ += num
            answer.append(summ)
```

```python
nums = [10,12,13,14]
sol = Solution()
sol.minElement(nums)
```
