```python
class Solution(object):
    def findDegrees(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: List[int]
        """
        result = []
        for i in range(len(matrix)):
            count = 0
            for j in range(len(matrix)):
                if (matrix[i][j] == 1):
                    count += 1
            result.append(count)
        return result
```

```python
matrix = [[0,1,1],[1,0,1],[1,1,0]]
sol = Solution()
sol.findDegrees(matrix)
```
