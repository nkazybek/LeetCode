```python
class Solution(object):
    def numIdenticalPairs(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        result = 0
        for i in range(len(nums) - 1):
            for j in range(i + 1, len(nums)):
                if (nums[i] == nums[j]):
                    result += 1
        return result
        
```

```python
nums = [1,2,3,1,1,3]
sol = Solution()
sol.numIdenticalPairs(nums)
```
