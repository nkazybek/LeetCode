```python
class Solution(object):
    def minBitFlips(self, start, goal):
        """
        :type start: int
        :type goal: int
        :rtype: int
        """
        start = str(bin(start)[2:])
        goal = str(bin(goal)[2:])
        newGoal = ""
        newStart = ""
        if len(start) > len(goal):
            
            for i in range(len(start)-len(goal)):
                newGoal += "0"
            newGoal += goal
            newStart = start
        else:
            for i in range(len(goal) -  len(start)):
                newStart += "0"
            newStart += start
            newGoal = goal
        answer = 0
        for i in range(len(newStart)):
            if newStart[i] != newGoal[i]:
                answer += 1
                
        return answer
        
```

```python
start = 10
goal = 7
sol = Solution()
sol.minBitFlips(start, goal)
```
