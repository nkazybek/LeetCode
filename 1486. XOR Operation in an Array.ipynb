```python
class Solution(object):
    def xorOperation(self, n, start):
        """
        :type n: int
        :type start: int
        :rtype: int
        """
        nums = []
        answer = 0
        for i in range(n):
            nums.append(start + 2 * i)
            answer = answer ^ nums[i]
        return answer
```

```python
n = 5
start = 0
sol = Solution()
sol.xorOperation(n, start)
```
