```python
class Solution(object):
    def mirrorDistance(self, n):
        """
        :type n: int
        :rtype: int
        """
        tmp = n
        reverse = 0
        while tmp > 9:
            lastDigit = tmp % 10
            reverse = reverse + lastDigit
            reverse *= 10
            tmp = tmp // 10
        reverse += tmp
        
        return abs(n - reverse)
```

```python
n = 25
sol = Solution()
sol.mirrorDistance(n)
```
