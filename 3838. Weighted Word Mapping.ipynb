```python
class Solution(object):
    def mapWordWeights(self, words, weights):
        """
        :type words: List[str]
        :type weights: List[int]
        :rtype: str
        """
        result = ""
        
        for string in words:
            total = 0
            
            for char in string:
                total += weights[ord(char) - 97]
            total = 26 - (total % 26)
            
            result +=(chr(total + 96))
        return result
```

```python
words = ["abcd","def","xyz"]
weights = [5,3,12,14,1,2,3,2,10,6,6,9,7,8,7,10,8,9,6,9,9,8,3,7,7,2]
sol = Solution()
sol.mapWordWeights(words, weights)
```
