```python
class Solution(object):
    def countPairs(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        answer = 0
        for i in range(len(nums) - 1):
            for j in range(i + 1, len(nums)):
                if nums[i] + nums[j] < target:
                    answer += 1
                    
        return answer
```

```python
nums = [-1,1,2,3,1]
target = 2
sol = Solution()
sol.countPairs(nums, target)
```
