```python
class Solution(object):
    def findPermutationDifference(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: int
        """
        answer = 0
        
        for i in range(len(s)):
            for j in range(len(t)):
                if s[i] == t[j]:
                    answer += abs(i - j)
                    
        return answer
```

```python
s = "abc"
t = "bac"
sol = Solution()
sol.findPermutationDifference(s, t)
```

```python
s = "abcde"
t = "edbac"
sol = Solution()
sol.findPermutationDifference(s, t)
```
