```python
class Solution(object):
    def recoverOrder(self, order, friends):
        """
        :type order: List[int]
        :type friends: List[int]
        :rtype: List[int]
        """
        result = [len(friends)]
        for i in order:
            for j in friends:
                print(i, j)
                if (i == j):
                    result.append(i)
        return result
```

```python
order = [3,1,2,5,4]
friends = [1,3,4]

solution = Solution()
print(solution.recoverOrder(order, friends))
```
