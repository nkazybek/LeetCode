```python
class Solution(object):
    def reverseDegree(self, s):
        """
        :type s: str
        :rtype: int
        """
        answer = 0
        for i in range(len(s)):
            value = 123 - ord(s[i])
            answer += value * (i + 1)
            
        return answer
```

```python
s = "abc"
sol = Solution()
sol.reverseDegree(s)
```

```python
s = "zaza"
sol = Solution()
sol.reverseDegree(s)
```
