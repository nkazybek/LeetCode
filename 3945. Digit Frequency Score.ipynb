```python
class Solution(object):
    def digitFrequencyScore(self, n):
        """
        :type n: int
        :rtype: int
        """
        result = 0
        while n > 9:
            tmp = n % 10
            n = n // 10
            result += tmp
        
        result += n
        return result
```

```python
n = 122
sol = Solution()
sol.digitFrequencyScore(n)
```

```python
n = 101
sol = Solution()
sol.digitFrequencyScore(n)
```
