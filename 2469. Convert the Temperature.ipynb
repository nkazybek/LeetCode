```python
class Solution(object):
    def convertTemperature(self, celsius):
        """
        :type celsius: float
        :rtype: List[float]
        """
        result = []
        result.append(celsius + 273.15)
        result.append(celsius * 1.80 + 32)
        return result
```

```python
celsius = 36.50
solution = Solution()
print(solution.convertTemperature(celsius))
```
