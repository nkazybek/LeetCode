```python
class Solution(object):
    def findWordsContaining(self, words, x):
        """
        :type words: List[str]
        :type x: str
        :rtype: List[int]
        """
        result = []
        added = False
        for i in range(len(words)):
            added = False
            for j in words[i]:
                if (added):
                    break
                elif (j == x):
                    result.append(i)
                    added = True

        return result
```

```python
words = ["leet","code"]
x = "e"
solution = Solution()
print(solution.findWordsContaining(words, x))
```

```python
words = ["abc","bcd","aaaa","cbc"]
x = "a"
solution = Solution()
print(solution.findWordsContaining(words, x))
```
