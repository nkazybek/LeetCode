```python
class Solution(object):
    def interpret(self, command):
        """
        :type command: str
        :rtype: str
        """
        answer = ""
        iterator = 0
        while (iterator < len(command)):
            if command[iterator] == "G":
                answer += "G"
            else:
                if command[iterator + 1] == ")":
                    answer += "o"
                    iterator += 1
                else:
                    answer += "al"
                    iterator += 3 
            iterator += 1
            
        return answer
```

```python
command = "G()(al)"
sol = Solution()
sol.interpret(command)
```

```python
command = "G()()()()(al)"
sol = Solution()
sol.interpret(command)
```

```python
command = "(al)G(al)()()G"
sol = Solution()
sol.interpret(command)
```
