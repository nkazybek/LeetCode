```python
class Solution(object):
    def alternatingSum(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        result = 0
        for i in range(len(nums)):
            if i % 2 == 0:
                result += nums[i]
            else:
                result -= nums[i]
                
        return result
        
```

```python
nums = [1,3,5,7]
sol = Solution()
sol.alternatingSum(nums)
```

```python
nums = [100]
sol = Solution()
sol.alternatingSum(nums)
```
