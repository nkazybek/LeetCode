```python
class Solution(object):
    def getSneakyNumbers(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        result = []
        nums = sorted(nums)
        for i in range(len(nums) - 1):
            if len(result) == 2:
                break
            if nums[i] == nums[i + 1]:
                result.append(nums[i])
        return result
```

```python
nums = [0,1,1,0]
sol = Solution()
sol.getSneakyNumbers(nums)
```

```python
nums = [0,3,2,1,3,2]
sol = Solution()
sol.getSneakyNumbers(nums)
```
