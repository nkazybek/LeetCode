```python
class Solution(object):
    def isValid(self, s):
        """
        :type s: str
        :rtype: bool
        """
        if len(s) % 2 != 0:
            return False
        stack = []
        
        for char in s:
            if char in "([{":
                stack.append(char)
            elif char in ")]}":
                if len(stack) > 0:
                    pop = stack.pop()
                else:
                    pop = ""
                if pop == "(" and char != ")":
                    return False
                if pop == "[" and char != "]":
                    return False
                if pop == "{" and char != "}":
                    return False
                if pop == "":
                    return False
        if len(stack) > 0:
            return False
        return True
        
```

```python
s = "()[]{}"

sol = Solution()
sol.isValid(s)
```

```python
s = "([])"

sol = Solution()
sol.isValid(s)
```

```python
s = "((])"

sol = Solution()
sol.isValid(s)
```

```python
s = "(("

sol = Solution()
sol.isValid(s)
```

```python
s = ")("

sol = Solution()
sol.isValid(s)
```

```python
s = "))"

sol = Solution()
sol.isValid(s)
```
