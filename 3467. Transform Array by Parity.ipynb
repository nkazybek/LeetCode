```python
class Solution(object):
    def transformArray(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        result = []
        even = 0
        for i in nums:
            if i % 2 == 0:
                even += 1
        for i in range(even):
            result.append(0)
        for i in range(len(nums) - even):
            result.append(1)
            
        return result
```

```python
nums = [4,3,2,1]
sol = Solution()
sol.transformArray(nums)
```
