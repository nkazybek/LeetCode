```python
class Solution(object):
    def maxFreqSum(self, s):
        """
        :type s: str
        :rtype: int
        """
        
        freq = {}
        
        for char in s:
            if char in freq:
                freq[char] += 1
            else:
                freq[char] = 1
        
        maxVowel = 0
        maxCons = 0
        for char in freq:
            if char in "aeiou":
                if freq[char] > maxVowel:
                    maxVowel = freq[char]
            else:
                if freq[char] > maxCons:
                    maxCons = freq[char]
                
        return maxVowel + maxCons
```

```python
s = "successes"
sol = Solution()
sol.maxFreqSum(s)
```

```python
s = "aeiaeia"
sol = Solution()
sol.maxFreqSum(s)
```
