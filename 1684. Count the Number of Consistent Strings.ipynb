```python
class Solution(object):
    def countConsistentStrings(self, allowed, words):
        """
        :type allowed: str
        :type words: List[str]
        :rtype: int
        """
        answer = 0
        for word in words:
            allow = True
            for char in word:
                if not char in allowed:
                    allow = False
            
```

```python
allowed = "ab"
words = ["ad","bd","aaab","baa","badab"]
sol = Solution()
sol.countConsistentStrings(allowed, words)
```
