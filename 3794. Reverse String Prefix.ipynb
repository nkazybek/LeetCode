```python
class Solution(object):
    def reversePrefix(self, s, k):
        """
        :type s: str
        :type k: int
        :rtype: str
        """
        answer = ""
        for i in range(k - 1, -1, -1):
            answer += s[i]
        for i in range(k, len(s)):
            answer += s[i]
            
        return answer
            
```

```python
s = "abcd"
k = 2
sol = Solution()
sol.reversePrefix(s, k)
```

```python
s = "xyz"
k = 3
sol = Solution()
sol.reversePrefix(s, k)
```
