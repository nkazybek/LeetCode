```python
class Solution(object):
    def minOperations(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: int
        """
        return sum(nums) % k
        
```

```python
nums = [3,9,7]
k = 5
solution = Solution()
solution.minOperations(nums, k)
```

```python
nums = [4,1,3]
k = 4
solution = Solution()
solution.minOperations(nums, k)
```
