```python
class Solution(object):
    def minimumOperations(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        result = 0
        for i in range(len(nums)):
            if (nums[i] % 3 == 0): continue
            else: result += 1
        return result
```

```python
nums = [1,2,3,4]
solution = Solution()
print(solution.minimumOperations(nums))
```
