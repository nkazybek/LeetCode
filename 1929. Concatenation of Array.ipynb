```python
class Solution(object):
    def getConcatenation(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        ans = []
        ans = nums + nums
        return ans
```

```python
nums = [1,2,1]

solution = Solution()
print(solution.getConcatenation(nums))
```
