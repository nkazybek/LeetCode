```python
class Solution(object):
    def convertDateToBinary(self, date):
        """
        :type date: str
        :rtype: str
        """
        answer = ""
        date = date.split("-")
        for i in date:
            answer += bin(int(i))[2:]
            answer += "-"
        return answer[:-1]
```

```python
date = "2080-02-29"
sol = Solution()
sol.convertDateToBinary(date)
```

```python
date = "1900-01-01"
sol = Solution()
sol.convertDateToBinary(date)
```
