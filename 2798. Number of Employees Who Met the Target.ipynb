```python
class Solution(object):
    def numberOfEmployeesWhoMetTarget(self, hours, target):
        """
        :type hours: List[int]
        :type target: int
        :rtype: int
        """
        answer = 0
        for hour in hours:
            if hour >= target:
                answer += 1
        return answer
```

```python
hours = [0,1,2,3,4]
target = 2
sol = Solution()
sol.numberOfEmployeesWhoMetTarget(hours, target)
```

```python
hours = [5,1,4,2,2]
target = 6
sol = Solution()
sol.numberOfEmployeesWhoMetTarget(hours, target)
```
